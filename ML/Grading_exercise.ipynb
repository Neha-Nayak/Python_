```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_excel("grading_.xlsx")
df
```

```python
df=df.drop(columns=['Unnamed: 4','Unnamed: 5','Unnamed: 6'])
df
```

```python
df1=df[0:40]
df1
```

```python
df['Grade'] = df['Grade'].replace('?', np.nan)
df_test=df[41:]
df_test
```

```python
df_testf=df_test.drop(columns=['USN','Name'])
df_testf
```

```python
from sklearn.preprocessing import LabelEncoder
le_grade=LabelEncoder()
```

```python
df1['grade'] = le_grade.fit_transform(df1['Grade'])
```

```python
df_main=df1.drop(['USN','Name','Grade'],axis='columns')
df_main
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df_main[['Percentage']],df_main.grade)
```

```python
reg.predict([[65]])         #0-distinction ,1-fail , 2-first , 3-second ,4-third
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('Percentage', fontsize=15)
plt.ylabel('grade', fontsize=15)
plt.scatter(df_main.Percentage,df_main.grade,color='blue',marker='*')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
model_dt=tree.DecisionTreeClassifier()
model_lr=LogisticRegression()
model_svm=SVC()
model_rf=RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_main[['Percentage']],df_main.grade,test_size=0.1)
```

```python
X_train
```

```python
def get_score(model, X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
get_score(LogisticRegression(),X_train,X_test,y_train,y_test)
```

```python
get_score(SVC(), X_train,X_test,y_train,y_test)
```

```python
get_score(tree.DecisionTreeClassifier(), X_train,X_test,y_train,y_test)
```

```python
get_score(RandomForestClassifier(), X_train,X_test,y_train,y_test)
```

```python
model_lr.fit(X_train,y_train)
model_lr.predict([[65]])  #0-distinction ,1-fail , 2-first , 3-second ,4-third
```

```python
model_svm.fit(X_train,y_train)
model_svm.predict([[65]])  # 0-distinction ,1-fail , 2-first , 3-second ,4-third
```

```python
model_rf.fit(X_train,y_train)
model_rf.predict([[65]])  # 0-distinction ,1-fail , 2-first , 3-second ,4-third
```

```python
model_dt.fit(X_train,y_train)
model_dt.predict([[65]])  # 0-distinction ,1-fail , 2-first , 3-second ,4-third
```

```python
for i in df_test['Percentage']:
    df_test.loc[df_test.Percentage == i, 'Grade']=model_lr.predict([[i]])
```
