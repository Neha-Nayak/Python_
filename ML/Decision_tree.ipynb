```python
import pandas as pd
```

```python
df = pd.read_csv("salary.csv")
df.head()
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_comp=LabelEncoder()
le_job=LabelEncoder()
le_deg=LabelEncoder()
```

```python
inputs['comp_n'] = le_comp.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['deg_n'] = le_deg.fit_transform(inputs['degree'])
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,0,1]])
```

```python
#exercise
df = pd.read_csv("titanic.csv")
df.head()
```

```python
dummy=pd.get_dummies(df.Sex)
dummy
```

```python
df1=pd.concat([df,dummy],axis='columns')
df1
```

```python
df2=df1.drop(['PassengerId','Name','Sex','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df2
```

```python
df2.Age=df2.Age.fillna(df2.Age.mean())
```

```python
inputs=df2.drop('Survived',axis='columns')
target=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_test
```

```python
model.predict([[3,7,29.1250,0,1]])
```
