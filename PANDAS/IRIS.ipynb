```python
import matplotlib.pyplot as plt
plt.subplot(2,2,1)
plt.title('IRIS-SepalLength')
bins=[3.5,4,4.5,5,5.5,6,6.5,7,7.5,8]
sl=df.SepalLength
plt.hist(sl,bins=bins)

plt.subplot(2,2,2)
plt.title('IRIS-SepalWidth')
bins=[1.5,2,2.5,3,3.5,4,4.5,5]
sw=df.SepalWidth
plt.hist(sw,bins=bins)

plt.subplot(2,2,3)
plt.title('IRIS-PetalLength')
bins=[1,2,3,4,5,6,7,8]
pl=df.PetalLength
plt.hist(pl,bins=bins)

plt.subplot(2,2,4)
plt.title('IRIS-PetalWidth')
bins=[0.5,1.0,1.25,1.5,1.75,2.0,2.5]
pw=df.PetalWidth
plt.hist(pw,bins=bins)

plt.tight_layout()
```
