```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv("iris_org.csv")
df1=df.sample(40)
df1
```

```python
g = df1.groupby("Species")
g
```

```python
data1=g.get_group("Iris-setosa")
data1
```

```python
sl1=data1.SepalLength
sw1=data1.SepalWidth
pl1=data1.PetalLength
pw1=data1.PetalWidth
print(sl1,sw1,pl1,pw1)
```

```python
plt.ylabel("PetalWidth")
plt.xlabel("SepalLength")
plt.title("Iris-setosa")
plt.bar(sl1,pw1,width=0.15)
plt.grid()
```
