```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos= np.arange(len(company)) #as plt cant take string value to plot the graph
xpos
```

```python
plt.xticks(xpos, company) #to get name as  x-parameter
plt.ylabel("Revenue(bln)")
plt.title("Stocks")
plt.bar(xpos,revenue,label="Revenue")

plt.legend()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln),Profit")
plt.xlabel("Company")
plt.title('Stocks')
plt.legend()
```

```python
plt.barh(xpos,revenue, label="Revenue") #for horizontal bar chart

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('Stocks')
plt.legend()
```
