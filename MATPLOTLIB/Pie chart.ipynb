```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels,explode=[0,0.2,0.1,0,0], radius=2,autopct='%1.2f%%',startangle=180)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,explode=[0,0.2,0.1,0,0], radius=2,autopct='%1.2f%%',startangle=180)
plt.savefig("piechart3.png",bbox_inches="tight",pad_inches=1, facecolor='w')
```
