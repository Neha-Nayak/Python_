```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
subs=[1,2,3,4,5,6]
max_m=[98,87,67,95,90,86]
min_m=[45,34,12,33,25,31]
avg_m=[71,39,66,64,57,58]
```

```python
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.title('Performance')
plt.plot(subs, max_m, label="max")
plt.plot(subs, min_m, label="min")
plt.plot(subs, avg_m, label="average")
plt.legend(loc='best',shadow=True)
plt.grid()
```
